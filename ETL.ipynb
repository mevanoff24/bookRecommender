# ETL (Extract, Transform, Load)

This notebook performs the Data Exploration and the ETL process(Extract, Transform, Load) to use for the machine learning recommender models. 

Observations and assumptions made in this ETL process
- We will be using only explicit ratings for building our book recommendation system.
- We keep outliers in the data (users with over 1000 total ratings) since we confirm that these ratings are not quantitatively different than the rest of the population. Although, we still do have a function in place to remove outliers outside three stds from the mean in the `BookDataSet` class. And we obviously test out both cases to see which results are 'better'
- Assign user with age under 5 and over 100 to NaN (will impute these in later phases)
- Remove image url data. A future project would be to also run a CNN to extract the image features to use as 'meta' data in something like a matrix factorization model. But I can't see this helping out too much for this book case. Maybe if we were to have an painting recommender system it would be useful. 

```python
import pandas as pd
import numpy as np
from pathlib import Path
from collections import Counter

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

DATA_PATH = Path('data/')
```

```python
def load_csv(filepath, **kwargs):
    '''
    input: (path to CSV file)
    read CSV file to pandas dataframe
    output: (pandas dataframe)
    '''
    return pd.read_csv(filepath, **kwargs)

def replace_slash(df):
    '''
    input: (pandas dataframe)
    Replaces '-' with '_' for column names
    output: (list)
    '''
    return [col.replace('-', '_') for col in df.columns]

ratings = load_csv(DATA_PATH/'BX-Book-Ratings.csv', sep=';', error_bad_lines=False, encoding='latin-1')
ratings.columns = replace_slash(ratings)
users = load_csv(DATA_PATH/'BX-Users.csv', sep=';', error_bad_lines=False, encoding='latin-1')
users.columns = replace_slash(users)
books = load_csv(DATA_PATH/'BX-Books.csv', sep=';', error_bad_lines=False, encoding='latin-1', low_memory=False)
books.columns = replace_slash(books)
```

The explicit ratings represented by 1â€“10 and implicit ratings represented by 0 will have to be segregated now. We will be using only explicit ratings for building our book recommendation system. Similarly, users are also segregated into those who rated explicitly and those whose implicit behavior was recorded.

```python
# only keep explicit ratings. All implicit ratings represented by 0 in data
ratings = ratings[ratings.Book_Rating != 0]
```

### Outlier detection - Should we remove these users? 

Lets see if the most frequent users mean value is statistically **higher** than the population in this dataset. 

Null: There is no statistical significance between the mean rating of the most frequent users and mean ratings of the population. 

Alternative: The mean rating of the most frequent users is quanitatility **higher** than the population from the mean ratings of the population. This mean is too high to be due to random chance.

Statistic: sample mean of ratings from users with over 1000 ratings

```python
def statistical_test(user_group, rating_count, iterations=10000):
    '''
    input: (
       user_group: pandas dataframe
       rating_count: int
       iterations: int
    )
    statistical test to test if the most frequent users mean value 
    is statistically higher than the population in this dataset.
    '''
    np.random.seed(100)
    over_rating_count = user_group[user_group['count'] >= rating_count]
    test_statistic = over_rating_count['mean'].mean()
    samples = []
    for _ in range(iterations):
        sample = user_group.sample(len(over_rating_count), replace=False)
        samples.append(sample['mean'].mean())
    samples = np.array(samples)
    p_val = len(samples[samples > test_statistic]) / len(samples)
    print('One sided P-Value: {}'.format(p_val))
    
    plt.hist(samples, bins=20)
    plt.axvline(test_statistic, c='red')
    plt.xlabel('Rating')
    plt.ylabel('Count')
    plt.title('Distribution of samples with test statitic line');
    
statistical_test(user_group, 1000, 10000)
```

The area in the tail beyond our observed statistic is 34.25%. Obviously, we are hereby declaring this tail area to be large, and therefore we will reject the alternative hypothesis. As the 'data' suggests that this sample *is* consistent with the overall population of ratings from users. Therefore we will keep these user in our dataset for now.

```python
def plot_location_counts(df, col, topk=20):
    '''
    input: (
       df: pandas dataframe
       col: pandas dataframe column name: str
       topk: int
    )
    plots a bar plot based on the counts of locations
    '''
    cols = list(locations_clean.columns)
    cols.remove(col)
    df.groupby(col).count().sort_values(cols[0], ascending=False)[cols[0]][:topk].plot(kind='barh')
    plt.title('{} Counts'.format(col))
```

```python
def factor_plot(df, x, y, kind='point', top_k=20):
    '''
    input: (
       df: pandas dataframe
       x: pandas dataframe column name: str
       y: pandas dataframe column name: str
       kind: type of seaborn plot: str
       topk: int
    )
    plots a seaborn catplot based on the columns passed 
    '''
    top = list(df[x].value_counts()[:top_k].index)
    sns.catplot(x, y, kind=kind,
                   data=df[df[x].isin(top)], height=5, aspect=10/2)
    plt.title('{} by {}'.format(y, x));
```
